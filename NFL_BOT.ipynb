```python
from scipy import stats

#For our regression line, use the string values for teams as the years (2010-2022)
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
y = [15, 14, 21.5, 14.4, 19.6, 18.5, 15.6, 18.4, 20.3, 19.2, 22.2, 21.9, 21.7]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
PPG_ALLOWED_DEF_Linear_Reg_Line =plt.show()
print(PPG_ALLOWED_DEF_Linear_Reg_Line)
#To make our compiler able to draw:
sys.stdout.flush()
```
